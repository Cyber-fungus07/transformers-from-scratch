```python
import tiktoken

# Load the exact tokenizer used by GPT-4
tokenizer = tiktoken.get_encoding("cl100k_base")
# Load the exact tokenizer used by GPT-2
tokenizer2 = tiktoken.get_encoding('gpt2')

print("Tokenizer loaded successfully!")
```

Output from the tokenizer used by GPT-4 :-

```python
token_id = tokenizer.encode("Transformerization")
print(f"Token IDs: {token_id}")
```

```python
for token in token_id:
  chunk = tokenizer.decode([token])
  print(f"{token} --> '{chunk}'")
```

Output of the Tokenizer used by GPT-2 :-

```python
novel_word = "Transformerization"


token_ids = tokenizer2.encode(novel_word)
print(f"Token IDs: {token_ids}")


for token_id in token_ids:
    chunk = tokenizer2.decode([token_id])
    print(f"ID {token_id} -> '{chunk}'")
```
